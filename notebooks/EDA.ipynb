```python
#criando colunas com as somatórias de vítimas por tipo de lesão

dados['lesoes_fatais'] = dados['Lesoes_Fatais_Tripulantes'] + dados['Lesoes_Fatais_Passageiros'] + dados['Lesoes_Fatais_Terceiros']
dados['lesoes_graves'] = dados['Lesoes_Graves_Tripulantes'] + dados['Lesoes_Graves_Passageiros'] + dados['Lesoes_Graves_Terceiros']
dados['lesoes_leves'] = dados['Lesoes_Leves_Tripulantes'] + dados['Lesoes_Leves_Passageiros'] + dados['Lesoes_Leves_Terceiros']
dados['ilesos'] = dados['Ilesos_Tripulantes'] + dados['Ilesos_Passageiros']
dados['lesoes_desconhecidas'] = dados['Lesoes_Desconhecidas_Tripulantes'] + dados['Lesoes_Desconhecidas_Passageiros'] + dados['Lesoes_Desconhecidas_Terceiros']
```

```python
#seleção das colunas relevantes para a análise e padronização de seus nomes

dados_v1 = dados[['Numero_da_Ocorrencia', 'Classificacao_da_Ocorrencia', 'Data_da_Ocorrencia', 'Municipio', 'UF', 'Regiao', 'Descricao_do_Tipo', 'Categoria_da_Aeronave', 'Fase_da_Operacao', 'Operacao', 'Danos_a_Aeronave', 'Modelo', 'Nome_do_Fabricante', 'lesoes_fatais', 'lesoes_graves', 'lesoes_leves', 'ilesos', 'lesoes_desconhecidas']]
dados_v1 = dados_v1.rename(columns={
    'Numero_da_Ocorrencia' : 'numero_da_ocorrencia',
    'Classificacao_da_Ocorrencia' : 'classificacao_da_ocorrencia',
    'Data_da_Ocorrencia' : 'data_da_ocorrencia',
    'Municipio' : 'municipio',
    'UF' : 'uf',
    'Regiao' : 'regiao',
    'Descricao_do_Tipo' : 'tipo_de_ocorrencia',
    'Categoria_da_Aeronave' : 'categoria_da_aeronave',
    'Fase_da_Operacao' : 'fase_da_operacao',
    'Operacao' : 'tipo_de_operacao',
    'Danos_a_Aeronave' : 'danos_a_aeronave',
    'Modelo' : 'modelo',
    'Nome_do_Fabricante' : 'nome_do_fabricante'
})
```

```python
#tratamento das colunas

dados_v1['numero_da_ocorrencia'] = dados_v1['numero_da_ocorrencia'].astype(str)
dados_v1['municipio'] = dados_v1['municipio'].str.upper()
dados_v1['fase_da_operacao'] = dados_v1['fase_da_operacao'].fillna('Não informado')
dados_v1['danos_a_aeronave'] = dados_v1['danos_a_aeronave'].fillna('Não informado')
dados_v1['data_da_ocorrencia'] = pd.to_datetime(dados_v1['data_da_ocorrencia'], format='%Y-%m-%d')
dados_v1['ano_da_ocorrencia'] = dados_v1['data_da_ocorrencia'].dt.year
dados_v1['lesoes_fatais'] = dados_v1['lesoes_fatais'].fillna(0).astype(int)
dados_v1['lesoes_graves'] - dados_v1['lesoes_graves'].fillna(0).astype(int)
dados_v1['lesoes_leves'] = dados_v1['lesoes_leves'].fillna(0).astype(int)
dados_v1['ilesos'] = dados_v1['ilesos'].fillna(0).astype(int)
dados_v1['lesoes_desconhecidas'] = dados_v1['lesoes_desconhecidas'].fillna(0).astype(int)
```
